# What is Numpy?
-NumPy stands for Numerical Python. It is a Python library used to work with arrays and perform fast mathematical operations on large amounts of data.

-It is faster than normal Python lists because it processes data in bulk using optimized code internally instead of looping element by element.

-NumPy is the foundation for most data science libraries — Pandas, Scikit-learn, and even deep learning libraries are built on top of it. That's why it's the first thing learned in any ML roadmap.

# Numpy Arrays Vs Python Sequences
1. Fixed Size
NumPy arrays have a fixed size once created. If you want to change the size, a new array is created and the old one is deleted. Python lists, on the other hand, can grow or shrink dynamically.

2. Same Data Type
All elements in a NumPy array must be of the same data type (e.g., all integers or all floats). Python lists can hold mixed data types (int, string, float together).

3. Faster Operations
NumPy arrays are much faster for mathematical operations because they are processed in bulk using optimized C code internally. Python lists require a loop to do the same task, which is slower.

4. Less Memory
NumPy arrays take less memory compared to Python lists because they store data in a more compact, fixed-type format.

5. Used in ML Libraries
Most data science and ML libraries (Pandas, Scikit-learn, TensorFlow) use NumPy arrays internally because of their speed and efficiency.

```python
# Array Functions
a1 = np.random.random((3,3))
a1 = np.round(a1*100)
a1
```

```python
## max/min/sum/prod - These functions work on the entire array by default. Using axis=0 applies the function column-wise, and axis=1 applies it row-wise.
# 0 -> col and 1 -> row
np.prod(a1, axis=0)
a = np.random.random((3, 3))
a = np.round(a * 100)
print(a)

print("Max:", np.max(a))
print("Min:", np.min(a))
print("Sum:", np.sum(a))
print("Prod:", np.prod(a))

print("Max col-wise:", np.max(a, axis=0))
print("Max row-wise:", np.max(a, axis=1))
```

```python
# mean/median/std/var - 
#mean - average of elements
#median - middle value when sorted
#std - standard deviation (spread of data)
#var - variance (square of standard deviation)
print("Mean:", np.mean(a))
print("Median:", np.median(a))
print("Std:", np.std(a))
print("Var:", np.var(a))
```

```python
# trigonometric functions - NumPy provides built-in trigonometric functions (sin, cos, tan) that apply element-wise on the array, just like scalar operations.
# trigonometric functions
print("Sin:\n", np.sin(a))
print("Cos:\n", np.cos(a))
print("Tan:\n", np.tan(a))
np.sin(a1)
```

```python
# dot product - Dot product is matrix multiplication between two arrays. For this, the number of columns in the first array must match the number of rows in the second array.
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(12,24).reshape(4,3)
a2 = np.arange(12).reshape(3, 4)
b2 = np.arange(12, 24).reshape(4, 3)

print(np.dot(a2, b2))
```

```python
# log and exponenets - np.log() calculates the natural logarithm of each element. np.exp() calculates e (Euler's number) raised to the power of 
#each element.
np.exp(a1)
a3 = np.arange(1, 11)

print("Log:\n", np.log(a3))
print("Exp:\n", np.exp(a3))
```

```python
# round/floor/ceil- round() rounds to the nearest value. floor() always rounds down to the smaller integer. ceil() always rounds up to the 
#larger integer.
np.ceil(np.random.random((2,3))*100)
a4 = np.random.random((2, 3)) * 100
print(a4)

print("Round:\n", np.round(a4))
print("Floor:\n", np.floor(a4))
print("Ceil:\n", np.ceil(a4))
```

```python
# Stacking - horizontal and vertical

#Stacking combines two or more arrays together. np.hstack() joins arrays side by side (horizontally) — number of rows must match. 
#np.vstack() joins arrays on top of each other (vertically) — number of columns must match.
#horizontal stacking 
a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(4,3)

a5
```

```python
# stacking
a = np.arange(12).reshape(3, 4)
b = np.arange(12, 24).reshape(3, 4)

print("Horizontal stack:\n", np.hstack((a, b)))
print("Vertical stack:\n", np.vstack((a, b)))
```
