# np.sort 
Return a sorted copy of an array

```python
# code
import numpy as np
a = np.random.randint(1,100,15)
a
```

```python
b = np.random.randint(1,100,24).reshape(6,4)
b
```

```python
np.sort(a)[::1]
```

```python
np.sort(b,axis = 0)
```

# np.concatenate
numpy.concatenate() function concatenate a sequence of arrays along an existing axis.

Concatenate means to link, join, or chain things together in a series or sequence.

```python
# code
c = np.arange(6).reshape(2,3)
d = np.arange(6,12).reshape(2,3)
print(c)
print(d)
```

```python
np.concatenate((c,d), axis = 1)
```

```python
np.concatenate((c,d), axis = 0)
```

# np.where
the numpy.where() function return the indices (index position) of elements in an input array where the given condition is satisfied.

```python
a
```

```python
# find all indices with value greater than 50
np.where(a>50)
```

```python
# replace all the values greater than 50 with 0
np.where(a>50,0,a)
```

```python
np.where(a%2 == 0,0,a)
```

# np.corrcoef
Return pearson product - moment correlation coefficient.

```python
salary = np.array([20000,40000,25000,35000,60000])
experience = np.array([1,3,2,4,2])
np.corrcoef(salary,experience)
```

# Set Functions
- np.union1d

- np.intersect1d

- np.setdiff1d

- np.setxor1d

- np.in1d

```python
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])
np.union1d(m,n)
```

```python
np.intersect1d(m,n)
```

```python
np.setdiff1d(m,n)
```

```python
np.setxor1d(m,n)
```

```python
m[np.in1d(m,1)]
```
