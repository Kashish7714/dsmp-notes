# What is Pandas
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python Programming language.


# Pandas Series
A Pandas series is like column in a table. It is  a 1D Array holding any type of data.

# Importing Pandas

```python
import numpy as np
import pandas as pd
```

# Series From Lists

```python

# string
country = ['India', 'Pakistan', 'USA', 'Srilanka']
pd.Series(country)
```

```python
# integers
runs = [13, 15, 14 , 78, 98]
run_ser = pd.Series(runs)
```

```python
# custom index 
marks = [23, 56,89,98]
subjects = [ 'maths', 'science', 'englisg', 'hindi']
pd.Series(marks, index = subjects)
```

```python
# Setting a name
marks = pd.Series(marks, index = subjects, name = ' Nitish ke marks')
marks
```

# Series from Dictionary

```python
marks = {
    'maths': 37,
     'Science' : 67,
    'hindi' : 98,
    ' english' : 78
}
marks_series = pd.Series(marks, name = ' Nitish ke marks')
marks_series
```
